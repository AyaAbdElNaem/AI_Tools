```python
def smiles_to_graph(smiles_str):
    # Parse the SMILES string into an RDKit molecule object
    mol = Chem.MolFromSmiles(smiles_str)
    if mol is None:
        return None

    # 1. Extract Node Features (Atoms)
    node_features = []
    for atom in mol.GetAtoms():
        # Feature vector: [Atomic Number, Is Aromatic (0 or 1)]
        features = [atom.GetAtomicNum(), int(atom.GetIsAromatic())]
        node_features.append(features)
    x = torch.tensor(node_features, dtype=torch.float)

    # 2. Extract Edge Indices and Features (Bonds)
    edge_indices = []
    for bond in mol.GetBonds():
        start, end = bond.GetBeginAtomIdx(), bond.GetEndAtomIdx()
        # PyG requires bidirectional edges for undirected molecular graphs
        edge_indices.append([start, end])
        edge_indices.append([end, start])

    edge_index = torch.tensor(edge_indices, dtype=torch.long).t().contiguous()

    # Create the PyTorch Geometric Data object
    graph_data = Data(x=x, edge_index=edge_index)
    return graph_data

# Example using the first compound from your dataset
example_smiles = "O=C1c2c(c3c4ccc(O)cc4n(C4OC(CO)C(O)C(O)C4O)c3c3[nH]c4cc(O)ccc4c23)C(=O)N1NC(CO)CO"
graph = smiles_to_graph(example_smiles)
print(graph)                                      # Output: Data(x=[num_atoms, num_features], edge_index=[2, num_edges*2])
```

```python
import pandas as pd

df = pd.read_csv('/content/kiba_with_official_names.csv')

# Clean dataset by dropping rows with missing essential values
df = df.dropna(subset=['compound_iso_smiles', 'target_sequence', 'affinity']).reset_index(drop=True)

# Display the first few rows and summary to verify
print(f"Dataset successfully loaded! Total rows: {len(df)}")
print(df.head())
```

```python
def smiles_to_graph(smiles_str):
    """Converts a SMILES string into a PyTorch Geometric Data graph object."""
    mol = Chem.MolFromSmiles(smiles_str)
    if mol is None:
        return None

    # 1. Extract Node Features (Atoms)
    node_features = []
    for atom in mol.GetAtoms():
        # Feature vector: [Atomic Number, Is Aromatic (0 or 1)]
        features = [atom.GetAtomicNum(), int(atom.GetIsAromatic())]
        node_features.append(features)
    x = torch.tensor(node_features, dtype=torch.float)

    # 2. Extract Edge Indices (Bonds)
    edge_indices = []
    for bond in mol.GetBonds():
        start, end = bond.GetBeginAtomIdx(), bond.GetEndAtomIdx()
        # Bidirectional edges for undirected graph
        edge_indices.append([start, end])
        edge_indices.append([end, start])

    if len(edge_indices) == 0:
        return None

    edge_index = torch.tensor(edge_indices, dtype=torch.long).t().contiguous()

    # Pack into a PyTorch Geometric Data structure
    return Data(x=x, edge_index=edge_index)

# Process all SMILES in the dataframe and store them in a Python list
print("Converting SMILES to molecular graphs...")
graph_objects_list = [smiles_to_graph(s) for s in df['compound_iso_smiles']]

# Note: Complex Python objects like custom graphs cannot be properly stored
# in plain Excel cell strings, so they are maintained in this execution list.
print(f"Successfully generated {len([g for g in graph_objects_list if g is not None])} valid graphs.")
```
